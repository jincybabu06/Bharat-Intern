```python
df['Pclass'].fillna(df['Pclass'].mode()[0],inplace=True)
df['Age'].fillna(df['Age'].mean(),inplace=True)
df['Cabin'].fillna(df['Cabin'].mode()[0],inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0],inplace=True)
df.isna().sum()
```

```python
df.drop(['Name'],axis=1,inplace=True)
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
df['Sex']=lb.fit_transform(df['Sex'])
df['Embarked']=lb.fit_transform(df['Embarked'])
df['Ticket']=lb.fit_transform(df['Ticket'])
df['Cabin']=lb.fit_transform(df['Cabin'])
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
df.drop(columns=['PassengerId','Ticket','Cabin','Fare'],axis=1,inplace=True)
```

```python
x = df.drop(['Survived'],axis=1)
x
```

```python
y = df['Survived']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(x_train)
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
x_train
x_test
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=7)
knn.fit(x_train,y_train)
y_pred=knn.predict(x_test)
y_pred
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
print("Score is",accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```
